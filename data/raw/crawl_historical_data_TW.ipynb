```python
import requests
import pandas as pd
import time
import datetime

SDate = "2024-01-01"
EDate = "2025-10-01"

# Chuyển ngày sang UNIX timestamp
# strptime(): (2025, 10, 7, 0, 0)
# timetuple(): (tm_year=2025, tm_mon=10, tm_mday=7, ...)
# mktime(): 1759795200 (số giây kể từ 1/1/1970)
start = int(time.mktime(datetime.datetime.strptime(SDate, "%Y-%m-%d").timetuple()))
end = int(time.mktime(datetime.datetime.strptime(EDate, "%Y-%m-%d").timetuple()))


tickers = [
    "2330.TW", "2317.TW", "2412.TW", "6505.TW", "2303.TW", "2881.TW", "2882.TW", 
    "1101.TW", "1301.TW", "2002.TW", "2337.TW", "2454.TW", "2603.TW", "2610.TW", 
    "3045.TW", "3037.TW", "3036.TW", "9933.TW", "3702.TW", "2886.TW", "2474.TW", 
    "2324.TW", "2357.TW", "3008.TW", "2395.TW", "2344.TW", "1723.TW", "6415.TW", 
    "1321.TW", "2449.TW", "2379.TW", "2327.TW", "1102.TW", "2412.TW", "9930.TW", 
    "2385.TW", "2891.TW", "2727.TW", "6269.TW", "4919.TW", "2408.TW", "2313.TW", 
    "2413.TW", "2375.TW", "8201.TW", "1504.TW", "2383.TW", "2707.TW", "3231.TW", 
    "3481.TW", "9910.TW", "5388.TW", "2301.TW", "2369.TW", "2356.TW", "2347.TW", 
    "2609.TW", "3443.TW", "2458.TW", "2368.TW", "3209.TW", "6414.TW", "2884.TW", 
    "2439.TW", "2006.TW", "3035.TW", "8081.TW", "2382.TW", "3038.TW", "3034.TW",  
]

headers = {
   "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                "AppleWebKit/537.36 (KHTML, like Gecko) "
                "Chrome/123.0.0.0 Safari/537.36"
}

all_ticker = []

for ticker in tickers:
    url = f"https://query1.finance.yahoo.com/v8/finance/chart/{ticker}?period1={start}&period2={end}&interval=1d"

    try:
        resp = requests.get(url, headers=headers)

        data = resp.json()
        result = data["chart"]["result"][0]
        timestamps = result["timestamp"]
        indicators = result["indicators"]["quote"][0]

        df = pd.DataFrame({
            "Date": pd.to_datetime(timestamps, unit="s"),
            "Open": indicators["open"],
            "High": indicators["high"],
            "Low": indicators["low"],
            "Close": indicators["close"],
            "Volume": indicators["volume"],
            "Ticker": ticker
        })

        all_ticker.append(df)
        print(f"✅ Hoàn tất {ticker}, {len(df)} dòng.")

    except Exception as e:
        print(f"error: {e}")

    time.sleep(2)  

# Gộp tất cả
if all_ticker:
    final_df = pd.concat(all_ticker, ignore_index=True)
    final_df.to_csv("historical_data_TW.csv", index=False)
    print("Đã lưu file historical_data_TW.csv thành công.")
else:
    print("Không có dữ liệu nào được lưu.")
```
